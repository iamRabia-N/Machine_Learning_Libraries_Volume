# 1. Introduction to Seaborn

<a id = "1.1"></a>
### 1.1. What is Seaborn
Seaborn is a popular data visualization library in python that helps data scientists and analysts create visually appealing and informative plots with ease. It is particularly well-suited for statistical data visualization.

<a id = "1.2"></a>
### 1.2. Why use Seaborn?
Seaborn provides an elegant and user-friendly way to create visually appealing data visualizations. It offers a high-level interface, aesthetically pleasing default styles and built-in themes that make it a preferred choice for many data analysts and scientists.

```python
pip install seaborn
```

# 2. Data Visualization Basics

<a id = "2.1"></a>
### 2.1. Importing Seaborn
Before you can start using Seaborn, you need to import it into your python script or jupyter notebook.

```python
import seaborn as sns
```

<a id = "2.2"></a>
### 2.2. Setting the Aesthetics
Seaborn comes with built-in themes and color palettes to enhance the appearance of your plots. You can set the style and context to control the aesthetics of your visualizations.

```python
sns.set_style("whitegrid")
sns.set_context("talk")
```

<a id = "2.3"></a>
### 2.3. Loading Datasets
Seaborn includes several sample datasets for practice. You can load these datasets using the `load_dataset()` function.

```python
tips = sns.load_dataset("tips")
```

<a id = "2.4"></a>
### 2.4. Plotting with Seaborn
Seaborn offers various functions for creating different types of plots. The fundamental plot types are univariate plots, bivariate plots and multivariate plots.

<a id = "3"></a>
# 3. Univariate Plots

<a id = "3.1"></a>
### 3.1. Understanding Univariate Plots
Univariate plots visualize a single variable. Here are some common types:

<a id = "3.2"></a>
### 3.2. Histogram
Histogram divides the data into discrete intervals. It counts the number of data points in each interval and represents this count with bars.

```python
sns.histplot(data=tips, x="total_bill")
```

<a id = "3.3"></a>
### 3.3. Kernel Density Estimation (KDE) Plot
KDE plot provides a smooth and continuous estimate of the data distribution. It creates a continuous curve that represents the probability density of the data.

```python
sns.kdeplot(data=tips, x="total_bill")
```

<a id = "3.4"></a>
### 3.4. Box Plot
Box plot displays the distribution of a dataset. It shows the median (middle value), quartiles (dividing the data into quarters) and potential outliers.

```python
sns.boxplot(data=tips, x="day", y="total_bill")
```

<a id = "3.5"></a>
### 3.5. Violin Plot
A violin plot combines key elements of a box plot and a kernel density plot. It displays the distribution of data, showing the range, median and quartiles like a box plot. Violin plot is useful for visualizing both summary statistics and the full data distribution simultaneously.

```python
sns.violinplot(data=tips, x="day", y="total_bill")
```

<a id = "4"></a>
# 4. Bivariate Plots

<a id = "4.1"></a>
### 4.1. Understanding Bivariate Plots
Bivariate plots visualize relationships between two variables. Here are some common types:

<a id = "4.2"></a>
### 4.2. Scatter Plot
Scatter plot shows the relationship between two continuous variables. It shows individual data points as dots. Typically used for displaying unconnected data points.

```python
sns.scatterplot(data=tips, x="total_bill", y="tip")
```

<a id = "4.3"></a>
### 4.3. Line Plot
Line plot is useful for visualizing trends over time or other continuous variables. It connects data points with lines and shows the trend or sequence of data over a continuous range.

```python
sns.lineplot(data=tips, x="day", y="total_bill")
```

<a id = "4.4"></a>
### 4.4. Pair Plot
Pair plot creates scatter plots for all combinations of variables in a dataset. It displays pairwise relationships between variables in a dataset and helps identify patterns and correlations between different features.

```python
sns.pairplot(data=tips)
```

<a id = "5"></a>
# 5. Multivariate Plots


<a id = "5.1"></a>
### 5.1. Understanding Multivariate Plots
Multivariate plots involve multiple variables. Here are some common types:

<a id = "5.2"></a>
### 5.2. Joint Plot
Joint plot combines multiple univariate and bivariate plots in a single figure. It often includes a scatter plot with marginal histograms or density plots.

```python
sns.jointplot(data=tips, x="total_bill", y="tip", kind="reg")
```

<a id = "5.3"></a>
### 5.3. PairGrid
PairGrid allows you to create customized pair plots for exploring relationships between multiple variables. It shows scatter plots, histograms, or other plots for pairs of variables. It is useful for visualizing relationships between multiple variables at once.

```python
import matplotlib.pyplot as plt

g = sns.PairGrid(data=tips, vars=["total_bill", "tip", "size"])
g.map(plt.scatter)
```

# 6. Customizing Plots

<a id = "6.1"></a>
### 6.1. Colors and Palettes
Seaborn provides various color palettes to customize your plots. You can set a palette using `sns.set_palette()`.

```python
sns.set_palette("Set2")   # "Set2" refers to one of the predefined color palettes in Seaborn
```

<a id = "6.2"></a>
### 6.2. Styling Plots
You can customize plot styles with `sns.set()`

```python
sns.set(style="whitegrid", rc={"grid.linewidth": 1.5}) 
```

<a id = "6.3"></a>
### 6.3. Axis Labels and Titles
Adding labels and titles to your plots is crucial for clarity. Use `plt.xlabel()` and `plt.title()`

```python
plt.xlabel("X-axis label")
plt.title("Plot Title")
```

<a id = "6.4"></a>
### 6.4. Annotations and Text
You can annotate your plots with text using `plt.text()`

```python
plt.text(2, 1, "Important point", fontsize=12, color='red')
```

# 7. Statistical Visualization

<a id = "7.1"></a>
### 7.1. Understanding Statistical Visualization
Seaborn's statistical plots are designed to reveal relationships in data. Let's explore some types.

<a id = "7.2"></a>
### 7.2. Distribution Plot
Distribution plot displays the distribution of a single variable. It is used for continuous data. Histograms, KDE plots, box plots, etc., are common types.

```python
sns.displot(data=tips, x="total_bill", kde=True)
```

<a id = "7.3"></a>
### 7.3. Categorical Plot
Categorical plot displays the distribution of data across categories. It shows how data is distributed within different categories or how one categorical variable relates to another. Bar plots, count plots, point plots, etc. are common types.

```python
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips") # Load a sample dataset from Seaborn
sns.catplot(data=tips, x="day", kind="count") # Create a categorical plot using sns.catplot
plt.show()
```

<a id = "7.4"></a>
### 7.4. Regression Plot
Regression plot helps you explore relationships between variables and fit regression models to your data.

```python
sns.lmplot(x="total_bill", y="tip", data=tips)
```

# 8. Advanced Seaborn Features

<a id = "8.1"></a>
### 8.1. FacetGrid
FacetGrid is a grid of subplots that allows you to create multiple plots for subsets of your data. It is useful for visualizing relationships within different categories or subsets of your data.

```python
g = sns.FacetGrid(tips, col="time", row="smoker")
g.map(sns.scatterplot, "total_bill", "tip")
```

<a id = "8.3"></a>
### 8.2. Heatmap
Heatmap is used to visualize matrix-like data, often showing correlations between variables. It is a specific type of plot that displays data values using a color scale.

```python
import matplotlib.pyplot as plt
import pandas as pd

# A DataFrame 
data = {
    'total_bill': [16.99, 10.34, 21.01, 23.68, 24.59],
    'tip': [1.01, 1.66, 3.50, 3.31, 3.61],
    'size': [2, 3, 3, 2, 4]
}

tips = pd.DataFrame(data)

correlation_matrix = tips.corr() 

plt.figure(figsize=(8, 6))  # Adjust the figure size if needed
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=0.5)

plt.show()
```
